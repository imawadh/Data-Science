```python
df = pd.DataFrame()
```

```python
df
```

```python
for i in range(1,50):
    response = requests.get(url=f"https://api.themoviedb.org/3/search/movie?api_key=3f35f2341eab85dd090c1dd3cdb78f8d&query={i}")
    #https://api.themoviedb.org/3/search/movie?api_key=3f35f2341eab85dd090c1dd3cdb78f8d&query=1
    print(response.json())
    curr_df = pd.DataFrame(response.json()['results'])
    curr_df = curr_df[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = pd.concat([df, curr_df],ignore_index=True)
```
