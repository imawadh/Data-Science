# Some Imputation Missing Techniques
## 1 -> Mean Inmputation Techniques

```python
sns.histplot(df['age'])
```

```python
sns.histplot(df['age'],kde=True)
```

```python
df['Mean Age Imputed']=df['age'].fillna(df['age'].mean())
```

```python
df['age']
```

```python
df[['Mean Age Imputed','age']]
```

```python
# mean imputation technique works with normally distributed values
```

# Miising Values Impuation techniques
# 2-> Meadian techniques

```python
df['Age_Median'] = df['age'].fillna(df['age'].median())
```

```python
df[['Age_Median','age','Mean Age Imputed']]
```

```python
# when we have outliers we need to use median because it is not sensitive to outliers
# if we use mean in case of skewed data we will encounter biasesness beacuse mean is sentive to outlier 
```

```python
import numpy as np
import pandas as pd

df['random_deck'] = df['deck'].apply(lambda x: np.random.choice(df['deck'].dropna()) if pd.isnull(x) else x)
```
